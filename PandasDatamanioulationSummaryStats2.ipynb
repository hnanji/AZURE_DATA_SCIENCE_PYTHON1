# Applying functions

```python
# this is applying iteative tasks over a dataframe

def rating_function(x):  # this defines the function
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

```python
# this creates a new column called rating category in the movies data frame.
# we use the apply() and apply the function to the data frame

movies["rating_category"] = movies["rating"].apply(rating_function)
```

```python
movies.columns
```

```python
# The lambda function does exactly the same thing


movies["rcategory"] = movies["rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')
```
