# a more flexible way of using different aggregration functions

```python
group_cols = ['d_smoking','d_male']
agg_dict = {'bmi':['sum', 'mean','size'],'age':['mean', 'var', 'size']}
df.groupby(group_cols).agg(agg_dict)
```

# manipulating multiple indexes

```python
group_cols = ['d_smoking','d_male']
agg_dict = {'bmi':['sum', 'mean','size'],'age':['mean', 'var']}
res = df.groupby(group_cols).agg(agg_dict).astype(int)
# the .astype(int)  at the end changes values and return only the intergers
```

```python
res
```

# Let's display the output of each level and then concatenate both levels before setting it as the new column values:

```python
 
level0 = res.columns.get_level_values(0)
level0
```

```python
level1 = res.columns.get_level_values(1)
level1
```

```python
res.columns = level0 + '_' + level1
```

```python
res
```
