# Multiple lines

```python
# multiple lines copuld be plotted before calling the plt.show()
# lines are added in layers
x = range(1,5)
plt.plot(x, [xi*1.5 for xi in x]) # 1st line
plt.plot(x, [xi*3 for xi in x]) # 2nd line
plt.plot(x, [xi/3 for xi in x]) # 3rd line
plt.show()
```

```python
# several lines could be plotted by passing the x and y values as a list to the plot() call like so
# colours are given to the lines automatically
x = range(1,10)
plt.plot(x, [xi*1.5 for xi in x], x, [xi*2.5 for xi in x], x, [xi*3.5 for xi in x])
plt.show()
```

```python
# the multiple plot above is possible because the default hold property is enabled
# this is s declaration that presevres all previously plotted lines instead of replacing it with a new line
import matplotlib.pyplot as plt
plt.interactive(True)# enables the interactive mode
plt.hold(False) # empty window will pop up
plt.plot([1,2,3])
# the hold option doesnt work, so check the back end settings
```

```python
x = range(1,10)
x = range(1,10)
# the label argument is added to the plot() call directly
plt.plot(x, [xi*1.5 for xi in x], label = 'A')
plt.plot(x, [xi*3 for xi in x], label = 'b')
plt.plot(x, [xi/3 for xi in x], label = 'c')
plt.legend()
plt.show()
```

```python
x = range(1,10)
x = range(1,10)
# alternative way to specify legend
plt.plot(x, [xi*1.5 for xi in x])
plt.plot(x, [xi*3 for xi in x])
plt.plot(x, [xi/3 for xi in x])
# the legends could be added as a list argument to the legend() call
plt.legend(['Slow','fast','fastest']) # the oder of plotting the lines is import for the legend
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
x=np.arange(-3,3,0.001)
y= x**2
plt.plot(x,y, 'g') #g changes the line colour to green using a string
plt.plot(x,norm.pdf(x), 'r-') # r- means solid line
# calling the axis object
axes = plt.axes() 
axes.grid()
axes.set_xlim([-4,4])
axes.set_ylim([0,9])
#axes.set_xticks([-4,-3,-2,-1,0,1,2,3,4])
plt.xlabel('a')
plt.ylabel('b')
plt.legend(['graph1','graph2'])
#plt.savefig('fig')
plt.show();
```

```python
fig = plt.figure() # this returns a figure
ax= fig.add_subplot(111) # this function returns the axis instance
ax1= fig.add_subplot(211) # defining plot region as 2 columns and 1 row and plotting in areas 1 which is top
ax1.plot([1,2,3],[1,2,3]) # plotting the image in the 1st plot area
ax2=fig.add_subplot(212) # defining the second plot area
ax2.plot([2,3,5],[1,2,5])# plotting the image in the 2nd plot area
```

```python
# Multiple figures can also be drawn by calling different figures
# the first figure can be called using figure(1) and the second as figure(2)
```

```python
fig1 = plt.figure() 
ax1= fig1.add_subplot(211) 
ax1.plot([1,2,3],[1,2,3])
fig2 = plt.figure()
ax2= fig2.add_subplot(111)
ax2.plot([1,2,3],[1,2,3]);
plt.show()
```

# Additional Y or (x axes)

```python
x=np.arange(0.,np.e,0.01)
y1 = np.exp(-x)
y2 = np.log(x)
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y1)
ax1.set_ylabel('y value for exp -x')
ax2 = ax1.twinx()
ax2.plot(x,y2,'r')
ax2.set_xlim([0,np.e])
ax2.set_ylabel('y values for lnx')
ax2.set_xlabel('same values for both x')
plt.show()
```
