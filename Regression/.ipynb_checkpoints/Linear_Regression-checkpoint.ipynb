```python
# LR is all abouty fitting a curve and some function through a set of observations and using that fiunction to predict new valkues
#  that you have never seen

#  Linear regression uses Ordinary Least Square (OLS) methed. This method mimimises the squared error between eacch point and the fitten line
#  by minimisimng the variance of the data pointy and the line, we can findn the best fitting line
#  slope= correlation between the twoi variables times the sd of y/sd of x

#  Another way of fitting best fit line across the data is called the gradiuent descent method
#  the gradientg descent which works best for 3 dimensiona data , tries to follow the contors of the data
#  gradient desctent is generally good for high dimensional data

#  R SQUARED OR COEFFICIENT OF DETERMINATION-
#   this describes how well the line or moldel fits  the data, how well does the model or line captures that vartiation in y
# are we getting equal amounts of variation on equal sided of the line?
#  Rsquared = 1  -  (sum of squared error/sum of squared variation from the mean)
```

```python
#  making up data to compute linear regression models, page 138s


%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import*
pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = 100 -(pageSpeeds + np.random.normal(0,0.1,1000))*3
plt.scatter(pageSpeeds,purchaseAmount)
```

```python
#   we import the stats package from the scipy package
```

```python
from scipy import stats
#   the linear regression formula will will retrn all this (slope, intercept, r_value, P_value, std_err) from which different 
#  statistics can be obtained
slope, intercept, r_value, P_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
# this is how we get the R squared value , the r_value is used to get the R squared value
r_value ** 2
```

```python
#  the R squared value is close to 1 meaning the model is a good fit and we have the same amount of variancce on either
#  side of the line
```

```python
#   y= mx + c
#  putchaseAmount = pageSpeeds * slope + intercept
import matplotlib.pyplot as plt
def predict(x):
    return slope * x + intercept
fitLine = predict(pageSpeeds)  # here we define a predict function to calculate the fitted line
plt.scatter(pageSpeeds, purchaseAmount) # here we plot a scatter  grapp for the two attributes
plt.plot(pageSpeeds, fitLine, c='r')  # here we add a fitted line through the data points

#   the predict function can be used to predict future values of purchasdeAmount for given values of pageSpeed times slope plus intercept

```

```python
#   creating another data set

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import*
np.random.seed(2)
pageSpeeds = np.random.normal(3.0,1.0,1000)
purchaseAmount = np.random.normal(50,10.1,1000)/pageSpeeds
plt.scatter(pageSpeeds,purchaseAmount)
```

```python
#   this converts the two attributes to np arrays
x =np.array(pageSpeeds)
y=np.array(purchaseAmount)
p4 = np.poly1d(np.polyfit(x,y,4)) # 4 for a fourth degree polynomial for the function, this could be changed to 3 or 2 accordingly
```

```python
import matplotlib.pyplot as plt
xp=np.linspace(0,7,100)
plt.scatter(x,y)
plt.plot(xp,p4(xp), c='r')
plt.show()
```

```python
#  computing the r-squared

#  by taking the y and the predcicted values(px(x)) in the r2_score() function that we have in sklearn.matrics can be used  calculate rsquare


from sklearn.metrics import r2_score
r2 = r2_score(y,p4(x))
print(r2) # this code simply compares a set of observations tyo a set of prediction and computes r square for you
```

```python
#  note that on the graph, up to 4hours of study results to less than 30% test grade
#  this is not a poerfect linear relationship, so we can fit higher order polynomials
```

```python
#  numpy has the vandert function that can be used to fit polynomials of different degree

for deg in [1,2,3,4,5]:
    lr.fit(np.vander(x,deg + 1),y);
    y_lr = lr.predict(np.vander(x,deg + 1))
plt.plot(x,y_lr,label='degree ' + str(deg));
plt.legend(loc=2);
print(r2_score(y,y_lr))
plt.plot(x,y,'ok')


# this method of fitting the polynomial is giving an error
#  this is expected tp produce diffent fits with different R squarted values to comopare and pick the best

   
```

```python
#  scikit-learn has a fgunction to generate new feature matrix consisting odf all polynimial combinations witrh a degree less than or 
#  equal to that specidfied

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
x = df.hours_stdied[:,np.newaxis]
y = df.test_grade
degree = 4 # specifying the degree  which can be changes and the r squared value will change as well
model = make_pipeline(PolynomialFeatures(degree), lr) # specifying the model of the polynomial
model.fit(x,y)
plt.scatter(x,y,color='black') #plotting observed values
plt.plot(x, model.predict(x), color='green') # plotting fitted values
print('R squared using built in function',r2_score(y, model.predict(x))) # getting the r squared
```

```python
#  COMPUTE CORRELATION USING A FUNCTION

def correlation (x,y):
    import numpy as np
    x_div = x - np.mean(x)
    y_div = y - np.mean(y)
    x_norm = x_div/np.std(x)
    y_norm = y_div/np.std(y)
    return np.mean(x_norm *y_norm)
```

```python
# defining a general model as a function that returns y as a function of inpute variables x and default coefficients a0,a1,a2 etc


# y = a0 + a1x + a2X^2 a3x^3

# def model(x,a0 =__, a1=___,a2=___ ):
    # return a0 + (a1*___) + (a2*___*___)

def model(x,a0 = 3,  a1= 1,a2=0):
    return a0 + (a1*x) + (a2*x*x)

#generate array of x, then estimate y for specific values of x
import numpy as np

x = np.linspace(-10,10,12) # values from -10 to 10, 21 of them
y = model(x)
print(y)


# plot the results
import statsmodels as rs
fig = rs.plot_prediction(x, y)


```

```python
# CORRELATION
# in correlation, the two variables must be standardized
# if not standardized, the measure will be covariation- which measures the association between two variables that will be affected 
# by the scale of each of the variables
```

```python
# standardized variable as follows

def standardized(variable):
    return(variable-np.mean(variable))/np.std(variable)

dataset['age_std'] = standardized(dataset['AGE'])

```

```python
dataset.head()
```

```python
# covariance

def covariance(variable_1, variable_2, biase=0):
    observations = float(len(variable_1))
    return np.sum((variable_1-np.mean(variable_1)) * (variable_2 - np.sum(variable_2)))/(observations-min(bias,1))

```

```python
# correlation
def correlation(var1,var2,bias=0):
    return covariance(standardized(var1), standardized(var2),bias)

```

```python
import statsmodels.formula.api as smf

# First, we define our formula using a special syntax
# This says that core temperature is explained by age
formula = "core_temperature ~ age"

# Perform linear regression. This method takes care of
# the entire fitting procedure for us.
model = smf.ols(formula = formula, data = dataset).fit()
print(model)
```

```python
# trying changing the age and see how temperature changes

def estimate_temperature(age):
    # Model param[0] is the intercepts and param[1] is the slope
    return age * model.params[1] + model.params[0]

print("Estimate temperature from age")
print(estimate_temperature(age=0))  # we can vary different values for age here
```

```python
import statsmodels.formula.api as smf

# here we are fitting a different univariate models in a loop, extracting the Rsquared value for each

for feature in ["male", "age", "protein_content_of_last_meal", "body_fat_percentage"]:
    # Perform linear regression. This method takes care of
    # the entire fitting procedure for us.
    formula = "core_temperature ~ " + feature
    simple_model = smf.ols(formula = formula, data = dataset).fit()

    print(feature)
    print("R-squared:", simple_model.rsquared)
```

```python
# Multiple regression usinh just two variables that we found predictive of temperature

model = smf.ols(formula = "core_temperature ~ age + male", data = dataset).fit()

print("R-squared:", model.rsquared)
```

```python
# NB as hammid for help
import statsmodels.formula.api as smf
def predict(age, male):
    '''
    This converts given age and male values into a prediction from the model
    '''
    model = smf.ols(formula = "core_temperature ~ age + male", data = dataset).fit()
    # to make a prediction with statsmodels, we need to provide a dataframe
    # so create a dataframe with just the age and male variables
    import pandas as pd
    df = pd.DataFrame(dict(age=[age], male=[male]))
    return model.predict(df)
```
