# Descriptive or Summary statistics

```python
import pandas as pd
import numpy as np
 
#Create a Dictionary of series
d = {'Name':pd.Series(['Alisa','Bobby','Cathrine','Madonna','Rocky','Sebastian','Jaqluine',
   'Rahul','David','Andrew','Ajay','Teresa']),
   'Age':pd.Series([26,27,25,24,31,27,25,33,42,32,51,47]),
   'Score':pd.Series([89,87,67,55,47,72,76,79,44,92,99,69])}
 
#Create a DataFrame
df = pd.DataFrame(d)
print(df)

# use the describe function for summary statistics of numeric variable

print(df.describe())

# for categorical variables, include the option: include ='object'

print(df.describe(include=['object']))

# we can show discriptive statistics for all the colunms  by adding the option include = 'all'

print(df.describe(include='all'))

# # the mean function can be used as follows

# calculate arithmetic mean

df. mean() # mean of all numeric variables in df

# column mean of the dataframe
df.mean(axis=0) # axis = 0 will calculate the columns wise mean of the df
```

```python
print(df.describe(include='all'))
```

```python
# for categorical variables, include the option: include ='object'

print(df.describe())
```

```python
# for categorical variables, include the option: include ='object'

print(df.describe(include=['object']))
```

```python
df.mean(axis = 1) # axis  calculates the row mean
```

```python
# mean of a specific column/ same applies for median/ mode/ sd/ variance

# mean of the specific column
df.loc[:,("Age","Score")].mean()
```

# Groupby function- sum, count, calculate mean based on a certain group

```python
import pandas as pd
import numpy as np
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine',
            'Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine'],
    'Exam':['Semester 1','Semester 1','Semester 1','Semester 1','Semester 1','Semester 1',
            'Semester 2','Semester 2','Semester 2','Semester 2','Semester 2','Semester 2'],
     
    'Subject':['Mathematics','Mathematics','Mathematics','Science','Science','Science',
               'Mathematics','Mathematics','Mathematics','Science','Science','Science'],
   'Score':[62,47,55,74,31,77,85,63,42,67,89,81]}
 
df = pd.DataFrame(d,columns=['Name','Exam','Subject','Score'])
print(df)
```

```python
# calculate the mean score of all subjects for each student
```

```python
df['Score'].groupby([df['Name']]).mean()
```

```python
df['Score'].groupby([df['Name'],df['Exam']]).mean() # here the sum is grouped by name and semester

# the function at the end can  be changed to sum(), describe()
```

```python
df['Score'].groupby([df['Name'],df['Exam']]).sum()
```

```python
df['Score'].groupby([df['Name'],df['Exam']]).describe()
```

```python
# we can group the entire data frame by name and exam

df.groupby(['Name','Exam']).sum()
```

# Descriptive statistics of the group :

```python
df['Score'].groupby(df['Subject']).describe()
```

```python
# we can print results like so
print(df['Score'].groupby(df['Subject']).describe())
```

```python
# the results can be passed to an object
a = df['Score'].groupby(df['Subject']).describe()
```

# cross tabulation - two way

```python
import pandas as pd
import numpy as np
 
#Create a DataFrame
d = {
    'Name':['Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine',
            'Alisa','Bobby','Cathrine','Alisa','Bobby','Cathrine'],
    'Exam':['Semester 1','Semester 1','Semester 1','Semester 1','Semester 1','Semester 1',
            'Semester 2','Semester 2','Semester 2','Semester 2','Semester 2','Semester 2'],
     
    'Subject':['Mathematics','Mathematics','Mathematics','Science','Science','Science',
               'Mathematics','Mathematics','Mathematics','Science','Science','Science'],
   'Result':['Pass','Pass','Fail','Pass','Fail','Pass','Pass','Fail','Fail','Pass','Pass','Fail']}
 
df = pd.DataFrame(d,columns=['Name','Exam','Subject','Result'])
df
```

```python
pd.crosstab(df.Subject, df.Result, margins= True) # adding options margins= TRUE display row/col totals
```

 # cross tab- 3 way

```python
pd.crosstab([df.Subject, df.Exam],df.Result, margins=True)
```

# Indexing with .ix: 

```python
# .iloc [1:m, 1:n] –  is used to select or index rows based on their position from 1 to m rows and 1 to n columns


# used to index a df by both name and position
df.iloc[:2]  # or df.iloc[:2,]  (df.iloc[2:5]= 3rd and 4th row)
```

```python
# select the first two columns
df.iloc[:,:2]
```

```python
# select first 1st and 4th columns
 
df.iloc[:,[0,3]]
```

```python
# select value by row label and column label using loc
 
df.loc[[1,2,3,4,5],['Name','Score']]
```

```python
# iloc –  iloc is used for indexing or selecting based on position .i.e. by row number and column number

#loc –   loc is used for indexing or selecting based on name .i.e. by row name and column name

#ix – indexing can be done by both position and name using ix.
```
